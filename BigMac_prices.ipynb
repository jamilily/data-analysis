```python
bigmac['dollar_price'].mean()
#bigmac['dollar_price'].sum()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
bigmac['dollar_price'].mean().plot(x='date', y='dollar_price',grid=True, figsize=(8, 6))
plt.title('BigMac Prices through time')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.grid(which='minor', linestyle='dotted')
plt.xticks(rotation=90)
plt.style.use('bmh')
plt.show()
```
